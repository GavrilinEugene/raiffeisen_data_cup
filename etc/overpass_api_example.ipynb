```python
def get_landuse_data(coord, node_type = 'landuse', distance = 200):
    api_query = 'way(around: ' + str(distance) + ', ' +\
            str(coord[0]) + ',' + str(coord[1]) + \
            ')["' + node_type + '"];(._;>;); out body;'
    result = api.query(api_query) 

    list_landuse = []
    for way in result.ways:
        list_landuse.append(way.tags.get("landuse", "n/a"))
    if len(list_landuse)  == 0 :
        list_landuse.append('n/a')
        
    most_common,num_most_common = Counter(list_landuse).most_common(1)[0]
    return most_common
```
