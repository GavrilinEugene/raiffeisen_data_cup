```python
import overpy

api = overpy.Overpass()

# fetch all ways and nodes
result = api.query('way(around:500,59.8426101, 30.1771015)[landuse];(._;>;);out;')
```

```python
result = api.query("""
    way(around:500,59.8426101, 30.1771015)["landuse"];
    (._;>;);
    out body;
    """)
```

```python
for way in result.ways:
    print("Name: %s" % way.tags.get("name", "n/a"))
    print("  landuse: %s" % way.tags.get("landuse", "n/a"))
    print("  Nodes:")
    for node in way.nodes:
        print("    Lat: %f, Lon: %f" % (node.lat, node.lon))
```
